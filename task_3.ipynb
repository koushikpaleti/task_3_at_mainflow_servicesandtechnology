```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])
plt.title("SCATTER PLOT")
plt.xlabel("YEAR")
plt.ylabel("OWN")
plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("line graph")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```
